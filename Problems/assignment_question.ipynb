**Question 5:** Write a Python program to calculate the factorial of a number.

```python
a = int(input("Enter the number"))

def calc_fact(n):
    if(n==0):
        return 1
    else:
        return n*calc_fact(n-1)
    
print("The factoial of",a,"is",calc_fact(a))
```

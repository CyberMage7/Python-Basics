```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.plot(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Basic graph")
# plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.plot(x,y,color="green",linestyle=':', marker='o')
```

```python
x1=[1,2,3,4,5,6]
y1=[1,4,9,16,25,36]
x2=[2,4,6,8,10,12]
y2=[4,8,12,16,20,24]

plt.figure(figsize=(9,5))

plt.subplot(2,2,1)
plt.plot(x1,y1,color="red",marker="o",linestyle="dashed")
plt.title("Plot 1")
plt.xlabel("X1--->")
plt.ylabel("Y1--->")

plt.subplot(2,2,2)
plt.plot(x2,y2,color="blue",marker="o",linestyle=":")
plt.title("Plot 2")
plt.xlabel("X2--->")
plt.ylabel("Y2--->")
```

```python
categories=['A','B','C','D','E']
values=[12,34,14,24,20]
plt.bar(categories,values,color="pink")
plt.xlabel("Categories")
plt.ylabel("Values")
```

```python
data=[1,2,2,3,3,3,4,4,4,5,5,5,5]
plt.hist(data,bins=5,color="orange",edgecolor="black")
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.scatter(x,y,color="red")
```

```python
labels=['A','B','C','D','E']
values=[20,10,30,30,10]
color=["green","orange","lightgreen","aqua","yellowgreen"]
explode=[0.1,0,0,0,0]
plt.pie(values,labels=labels,explode=explode,colors=color,autopct="%1.1f%%",shadow="True")
```
